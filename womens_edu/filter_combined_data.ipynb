```python
dataset_root_dir = '/content/drive/MyDrive/Colab Notebooks/cs230/womens_edu/data/'
```

```python
combined_df = pd.read_csv(dataset_root_dir + '/combinedss.csv')
```

```python
combined_df.head() 
```

```python
street_imgs = ['AM', 'BJ', 'CD', 'CM', 'GH', 'KY', 'MD', 'NM', 'NP', 'TJ', 'ZW']
```

```python
filtered = combined_df[ combined_df['cc'].isin(street_imgs) ]
```

```python
street_img_count = filtered.groupby(['DHSID_EA'])['img_id'].count()
street_img_count.tail(10)
```

```python
sample_size = 5       
sample_fn = lambda x: x.sample(n=min(sample_size, len(x)))

filtered_sampled = filtered.groupby('DHSID_EA', as_index=False).apply(sample_fn)
filtered_sampled.head()
```

```python
street_img_count_sampled = filtered_sampled.groupby(['DHSID_EA'])['img_id'].count()
street_img_count_sampled.head(10)
```
